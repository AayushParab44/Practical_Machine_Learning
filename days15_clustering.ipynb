# K-Means

- In K - Means clustering , we choose an arbitrary centroid first, then with every iteration,  <br>
  the centroid gets updated until it comes to the correct centroid of the cluster.

- AS we increase the number of clusters, the variation within the clusters will keep going down(decrease).

- AS we increase the number of clusters, the variation between the clusters will increase.


Sampling Techniques-
- Random Sampling -
- Segmented odd sampling-
- Quota sampling - 
- snowball sampling- 

```python
import pandas as pd
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('CDAC_DataBook.xlsx',sheet_name='clusters')
```

```python
mod1= KMeans(n_clusters=3).fit(df)
```

```python
mod1.cluster_centers_
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df.Var1,df.Var2,c='r')
plt.scatter(29.6, 66.8,c='g')
plt.scatter(43.2, 16.7,c='b')
plt.scatter(55.1, 46.1,c='m')
```

```python
plt.scatter(df.Var1,df.Var2,c=mod1.labels_)
plt.scatter(29.6, 66.8,c='r')
plt.scatter(43.2, 16.7,c='r')
plt.scatter(55.1, 46.1,c='r')
```

```python
models=[]
clust_nums=range(3,16)
int_gain=[]
for i in range(3,16):
    mod=KMeans(n_clusters=i).fit(df)
    models.append(mod)
    int_gain.append(mod.inertia_)

plt.plot(clust_nums,int_gain)
```
