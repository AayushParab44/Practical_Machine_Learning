# Simple Exponential Smoothing

```python
from statsmodels.tsa.api import SimpleExpSmoothing
```

```python
mod1= SimpleExpSmoothing(np.asarray(df_train.BirthRate)).fit()
df_test['SES']=mod1.forecast(len(df_test))
rmse6= (mean_squared_error(df_test.BirthRate,df_test['SES']))**.5
rmse6
```

# ARIMA model- Auto Regressive Integrated Moving Average
- Arima model would be used ONLY when the time series is stationary(mean and variance constant).(similiar to sinusoidal curve)


```python
# to check whether the time series is stationary or not, we use the ADFuller test.
# Ho: TS is not stationary.
# Ha: TS is stationary.
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller(df_train)
```

p-value is 0.9723576777571009 >0.05 hence the time series is not stationary.

Method of differencing-
- to convert the time series which is not stationary to  a one which is stationary.
- differencing is finding the difference between consecutive numbers in the time series and then plotting those on the times series graph.
- in (p,d,q) the 3 parameters that we have to search for, d indicates the number of levels of differencing <br>
after which the time series becomes stationary.


```python
d1= df_train.diff().dropna()
```

```python
adfuller(d1)
```

p=0.0005333333<0.05, implies time series has now become stationary.

```python
d1.plot(kind='line')  #final pval<0.05
```
